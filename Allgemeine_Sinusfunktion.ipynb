# Allgemeine Sinusfunktion
\begin{equation}
y(t)=A*sin(\omega*t+\phi)
\end{equation}

## Amplitude

A...Amplitude, A e R

A>1... Streckung in y-Richtung

0<A<1...Stauchung in y-Richtung

A<0...Spiegelung an der x-Achse + Stauchung bzw. Streckung


## Kreisfrequenz

$\omega$ ...Kreisfrequenz; T...Periode(Umlaufzeit in einem Kreis)

$T=2\pi/\omega$


$\omega>1$ ...Periode wird verkürzt

$0<\omega<1$ ...Periode wird vergrößert

$\omega<0$ ... Cosinus


## Phasenwinkel

$\phi$ ...Nullphasenwinkel | $\omega*t+\phi$ ...Phasenwinkel

$\phi<0$ ...rechtsverschiebung auf der x-Achse

$\phi>0$ ...linksverschiebung auf der x-Achse

```python
import matplotlib.pyplot as plt
import numpy as np

def sinusplot(A, w, p):
    """
    A = Amplitude
    w = Winkelgeschwindigkeit
    p = Phasenwinkel
    """
    # Grundmenge
    t = np.arange(0.0, np.pi * 2., 0.01)

    # Formel
    y = A * np.sin(w *  t + p)

    # Plot
    fig, ax = plt.subplots()
    ax.plot(t, y)

    ax.set(
       xlabel='time(t)',
       ylabel='Amplitude(y)',
       title='Sinuskurve',)
    ax.grid()

    plt.show()
```

## bsp.

```python
sinusplot(A=1.3, w=2.4, p=np.pi/3.1)
```

## Achtung bei Nullstellen
$t_0=-\frac{\phi}{\omega}$

$\frac{\pi}{2}=\frac{\phi}{1}=-\phi$ ==) $\phi=-\frac{\pi}{2}$

--)--)

$\frac{\pi}{4}= -\frac{\phi}{2}$  |/2

$\frac{\pi}{2}=-\phi$ ==) $\phi=\frac{\pi}{2}$
